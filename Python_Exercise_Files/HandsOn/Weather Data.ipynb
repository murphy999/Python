```python
# Import Pandas Library
import pandas as pd

# Read CSV File
weather = pd.read_csv("waether.csv",header=None)

weather # Print DF
```

```python
# Add Headers to the data
weather.columns=["Date","City","Temperature","Windspeed","Event"]
```

```python
# Check For Null Values
weather.isnull().sum()
weather.isnull().sum().sum()
```

```python
# Data Imputation - Drop Records with Missing Values
weather.dropna(how = "all") # default value 'any' => any column value null for a record then drop
# ='all' => if all null values for a record then only drop

# Data Imputation - Substituting Missing Values with Mean Value 
weather = weather.fillna(weather[["Temperature","Windspeed"]].mean())

# Ref: https://towardsdatascience.com/6-different-ways-to-compensate-for-missing-values-data-imputation-with-examples-6022d9ca0779
```
